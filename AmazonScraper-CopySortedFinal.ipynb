```python
def get_product_info(page_no):

    r = requests.get(URL1+str(page_no)+URL2+str(page_no), headers = headers)#, proxies=proxies)
    #print(r.status_code)
    content = r.content #r.text? or r.content?
    soup = BeautifulSoup(content, 'lxml') #try html parser and not lxml
    #print(soup)
    
    data_list = []
    
    #try to find all: sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20 s-list-col-right instead of just a-section
    
    ratings_list = [] #a-spacing-small a-spacing-top-small 'add this to a-section?'
    num_ratings = [] #sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20 s-list-col-right
    stuff = soup.find_all('div', class_ = "sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20 s-list-col-right") #so some things print 3 times and im not really sure why
    #stuff = soup.findAll('div', attrs={'id':'a-page'})
    #print(len(stuff))
    #print(stuff)
    for i in stuff:
        
        temp_list = []
        
        name = ""
        temp = i.find('span', attrs = {'class':'a-size-medium a-color-base a-text-normal'})
        if temp != None:
            #if temp.text in name_list: #trying to ENSURE WE GET ONLY UNIQUE ENTRIES.
                #continue
            if temp.text not in name_list:
                #name_list.append(temp.text) #we'll just manually get them out. get the data for now.
                #print(temp.text)
                name = temp.text
                temp_list.append(name)
    
        rating = ""
        temp1 = i.find('span', attrs = {'class':'a-icon-alt'})
        if temp1 != None:
            if temp1.text not in name_list:
                ratings_list.append(temp1.text)
                #print(temp1.text)
                rating = temp1.text
                temp_list.append(rating)
        else:
            #print('NO RATING')
            rating = 'NO RATING'
            temp_list.append(rating)
             
        num_rating = ""
        temp2 = i.find('span', attrs = {'class':'a-size-base s-underline-text'})
        if temp2 != None:
            if temp2.text not in name_list:
                num_ratings.append(temp2.text)
                #print(temp2.text)
                num_rating = temp2.text
                temp_list.append(num_rating)
        else:
            #print('NO NUM RATING')
            num_rating = 'NO NUM RATING'
            temp_list.append(num_rating)
            
                
        #price var:
        price_string = ""
        ###PRICE INFO###
        temp3 = i.find('span', attrs = {'class':'a-price-symbol'})
        if temp3 != None:
            if temp3.text not in name_list:
                num_ratings.append(temp3.text)
                #print(temp3.text, end='')
                price_string += temp3.text
                
        temp3 = i.find('span', attrs = {'class':'a-price-whole'})
        if temp3 != None:
            if temp3.text not in name_list:
                num_ratings.append(temp3.text)
                #print(temp3.text, end='')
                price_string += temp3.text
                
        temp3 = i.find('span', attrs = {'class':'a-price-decimal'})
        if temp3 != None:
            if temp3.text not in name_list:
                num_ratings.append(temp3.text)
                #print(temp3.text, end='')
                price_string += temp3.text
                
        temp3 = i.find('span', attrs = {'class':'a-price-fraction'})
        if temp3 != None:
            if temp3.text not in name_list:
                num_ratings.append(temp3.text)
                #print(temp3.text)
                price_string += temp3.text
                
        #print('full price: ', price_string)
        temp_list.append(price_string)
             
        ######
        
        #comp_info = i.findAll('span', attrs = {'class':'a-text-bold'})[1].text
        #comp_info = [c for c in comp_info if c != []]
        #print(comp_info)
        
        #g = i.findAll('span', attrs = {'class':'a-color-secondary'})[0].text
        #print(g)
        
        comp_info = i.find('div', class_='sg-row s-product-specs-view')
        
        if comp_info != None:
            split_info = comp_info.find_all('span', attrs={'class':'a-text-bold'})
            #print()
            #print(split_info[0].text)
            temp_list.append(split_info[0].text)
            #print(split_info[1].text)
            temp_list.append(split_info[1].text)
            #print(split_info[2].text)
            temp_list.append(split_info[2].text)
            #print(split_info[3].text)
            temp_list.append(split_info[3].text)
            #print()
        else:
            temp_list.append('No Info')
            temp_list.append('No Info')
            temp_list.append('No Info')
            temp_list.append('No Info')
        
        #print('temp list: ', temp_list)
        data_list.append(temp_list)
        
    return data_list
```

```python
def make_gaming_dataframe(n_pages = 10):
    '''
    This function creates the initial dataframe of information from the amazon scrapper. Further processing
    will occur after calling this function and creating this dataframe in its raw form.
    
    args: (n_pages): Specifies how many amazon pages of information you wish to scrape (22 records per page).
    
    '''

    df = pd.DataFrame(columns = ['Product Name', 'Avg. Rating', '# Ratings', 'Price', 'Processor Type', 'Disk Size', 'RAM', 'Processor Speed'])
    num_pages = n_pages

    for j in range(2, num_pages+2):
        #print(j)
        new_info = get_product_info(j) #instead of j+1
        #print(new_info)

    
        temp_df = pd.DataFrame(columns = ['Product Name', 'Avg. Rating', '# Ratings', 'Price', 'Processor Type', 'Disk Size', 'RAM', 'Processor Speed']) 
        for i in range(len(new_info)):
        
            #print(len(new_info[i]))
            #adding to make sure length is 8 to match length of dataframe
            if len(new_info[i])==8:
            
                temp_df.at[i, :] = new_info[i]
                #temp_df.at[i, 'Page No.'] = j
        
        df = pd.concat([df, temp_df], ignore_index = True)
        print('Page: ', j)

    return df
```
